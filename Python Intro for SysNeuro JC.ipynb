## Print
Printing statements to the console can be helpful for checkpoints in your code. In python strings are created by place text between a pair either a ' or " symbols. You can also create formatted strings if you want to add variables into your string by starting with f' or f" , adding variables in place with { }, and closing with ' ". 

````
#Printing strings to output
print('Hello world')
print(f'The speed of light is {c} m/s')
````

You can also perform computations on variables inside of a string without modifying the variable.

````
i = 3
print(f'if i is {i}, then i squared is {i**2}')
print(f'but i is still {i}')
````

## For loops
For loops in python are similar to those in Matlab, with some syntax differences: the for loop iterates over members `in` a sequence in order, using a `:` at the end of the for statement, indentation matters. This is a simple Matlab for loop:
```
for i = 0:3
  disp(i)
end
```
Lets write it in Python:

````
for i in range(4):
  print(f'{i}')

test = 2
````

You can also iterate over members of a string, like this:

````
print('It\'s fun to stay at the')
for i in 'YMCA':
  print(f'{i}')
````

It can be really easy to make a one-off indexing mistake when switching between Matlab and Python. In Python the first index of a alphabetical sequence is 'b' not 'a', which is the zeroeth index. This is because Python counts the number of elements from the first element, rather than counting like a human being. 

````
a = 0
for i in range(4):
  a += 1
  print(a)
````

## Importing Libraries
While Python has a lot of powerful general tools, most of the time you will need to use **libraries**. A library is a collection of functions, classes, variables that implement a set of related tasks. To use these libraries you use the `import`function. Two of the most common libraries are numpy ([documentation](https://numpy.org/)), which is a tool for working with matrices and arrays, and Matplotlib ([documentation](https://matplotlib.org/)), which is a tool for making all kinds of plots. We also add the `as` statement so that we can access the library with this shortened name. 

````
import numpy as np
import matplotlib.pyplot as plt
````

#Working with arrays in Numpy

````
#Lets fabricate some data
data = np.arange(12).reshape(3,4)
#We can find out about the array by using built in atributes
print(data.shape) #(rows, columns)
print(data.ndim) #number of dimensions
print(data.dtype) #data type
print(data.size) #number of elements
print(data[2][3]) #select the last element of the array
````

````
#Some other common ways to create arrays
data = np.zeros((3,4)) #an array of shape (3,4) with zeros for all elements
data = np.ones((3,4)) #an array of shape (3,4) with ones for all elements 
data = np.eye(3) #a square array of shape (3,3) with ones on the diagonal and zeros else
print(data)
````

````
#You can perform some basic matrix operations in numpy
data = np.arange(9).reshape(3,3)
print(data)
````

````
data**2
````

````
#You can also do linear algebra with numpy
#Matrix Inverse
data = np.array([[1,2], [3,4]])
print(np.linalg.inv(data))

a = np.array([[3,1], [1,2]])
b = np.array([9,8])
c = np.linalg.solve(a, b) #solve the linear system of equations
print(c)
````

#Plotting with Matplotlib
